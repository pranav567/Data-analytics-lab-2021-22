```python
x_tr2 = x_tr.copy()
```

```python
for col in x_tr2.columns:
    col_zscore = col + "_zscore"
    x_tr2[col_zscore] = (x_tr2[col] - x_tr2[col].mean())/x_tr2[col].std(ddof=0)
x_tr2.head()
```

```python
x_tr2.shape
```

```python
for col in x_tr.columns:
  x_tr2 = x_tr2[abs(x_tr2[col+'_zscore'])<3]
x_tr2
```

```python
x_tr2.describe().T
```

```python
x_tr3 = x_tr2[x_tr.columns]
```
